* 클래스(Class) - 같은 종류(또는 문제 해결을 위한)의 집단에 속하는 속성(attribute)과 행위(behavior)를 정의한 것으로 객체지향 프로그램의 기본적인 사용자 정의 데이터형(user define data type)이라고 할 수 있다. 클래스는 프로그래머가 아니지만 해결해야 할 문제가 속하는 영역에 종사하는 사람이라면 사용할 수 있고, 다른 클래스 또는 외부 요소와 독립적으로 디자인하여야 한다.

* 메서드(Method) - 클래스로부터 생성된 객체를 사용하는 방법으로서 객체에 명령을 내리는 것이라 할 수 있다. 메서드는 한 객체의 속성을 조작하는 데 사용된다.

### class 메소드

메소드란 클래스 내의 함수로서 클래스 인스턴스에서 호출해서 사용합니다. 클래스 메서드에는 항상 첫번째 인자로 **self** 가 전달된다는 점을 명심하자.

### 클래스 생성자 매개변수

클래스에 인자를 가지게 선언해 보자

생성자 매개변수를 이용한 클래스 선언

````python
# 2개의 인자를 가진 클래스
class Point:
    def __init__(self, x, y):
        print("init constructor...", x, y)
        self.x = x
        self.y = y
    def __new__(self, x, y):
        print("new constructor...", x, y)
        return super().__new__(self)
````

````python
my_point = Point(1,2)  # 생성자 호출 (x, y)
my_point
````

초기화한 속성은 메서드에서 이용 가능하다. 초기화한 속성을 사용하거나 값을 할 당해서 사용한다.

````python
my_point.x, my_point.y
````

````python
my_point.x = 10.0
my_point.y = 20.0
print(my_point.x, my_point.y)
````

````python
print(my_point.x, my_point.y)
my_point.x = 1.0
my_point.y = 2.0
print(my_point.x, my_point.y)
````

### 접근제어

Python은 기본적으로 다른언어에 있는 접근제어자(public, private, protected)가 없다. 다만 의도적으로 아래의 코드에서 ` self._pi`  처럼 변수명의 `_` 는 내부적으로 사용되는 변수로 이해한다..

````python
class Point:
    def __init__(self, x, y):
        self._pi = 3.14
        self.x = 0.0
        self.y = 0.0
        return super().__init__()

    def __new__(self, x, y):
        return super().__new__(self)
````

````python
my_point = Point()
my_point._pi
````

````python
my_point._pi = 0

my_point._pi
````

### 클래스 속성과 객체 속성

원천적인 접근을 막으려면 `__` 더블 언더바를 사용하면 막을 수 있다.

````python
class Point:
    def __init__(self, x, y):
        self.__pi = 3.14
        self.x = 0.0
        self.y = 0.0
        return super().__init__()

    def __new__(self, x, y):
        return super().__new__(self)
    
    def pi(self):
        return self.__pi
````

````python
my_point = Point(1, 2)
my_point.pi()
````

````python
my_point.__pi = 0
my_point.pi()
````

`__` 를 통해서 클래스 속성으로 선언되고, 클래스 객체에서 같은 이름의 속성을 선언해도 별도의 다른 속성으로 다뤄진다

````python
my_point.__pi
````

### Validation

보통 객체 생성시 들어올 값에 대해서 __init__에서 Validation을 수행합니다.

````python
class Flight:

    def __init__(self, number):
        if not number[:2].isalpha():
            raise ValueError("첫 두글자가 알파벳이 아닙니다.")
        if not number[:2].isupper():
            raise ValueError("첫 두글자가 대문자가 아닙니다.")
        if not number[2:].isdigit():
            raise ValueError("세번째 글자 이상이 양의 숫자가 아닙니다.")
        self._number = number
````

### Overriding

같은 이름의 메서드를 다른 매개변수를 사용해서 새롭게 정의할 수 있다.

````python
class VIPMember(Member):
    # 생성자
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 메서드 오버라이딩
    def greeting(self):
        print('You are very ViP')
    
    def info(self):
        print('<VIP>', '이름:', self.name, ' 나이:', self.age)
````

````python
member1 = Member('홍길동', 20)
member1.greeting()
member1.info()
````

````python
vipmember = VIPMember('강감찬', 21)
vipmember.greeting()
vipmember.info()
````

## 연산자 오버로딩(Operator Overloading)

파이썬 객체와 객체의 연산을 위해 제공하는 특별한 메서드를 재정의 할 수 있다.



### `__add__`

````python
class Point:
    def __init__(self, x, y):
        self.x = 0.0
        self.y = 0.0
        self._pi = 3.14
    
    def __add__(self, num):
        self.x += float(num)
        self.y += float(num)

    def __sub__(self, num):
        self.x -= float(num)
        self.y += float(num)
````

````python
p1 = Point(10, 20)
p2 = Point(4, 9)
````

````python
p1 + 20
p1.x
````

````python
class Point: 
    def __init__(self, num): 
        self.num = num 

    def __add__(self, other): 
        self.num += other.num 
        print(self.num)

    def __sub__(self, num): 
        self.num -= num
````

````python
n = Point(40)
m = Point(40)
````

````python
n + m
````

### `__str__`


````python
class Point:
    def __init__(self, x, y):
        self.x = 0.0
        self.y = 0.0
        self._pi = 3.14
    
    def __str__(self):
        return "Point({0}, {1})".format(self.x, self.y)
````

````python
p1 = Point(10, 20)
````

````python
print(p1)
````
